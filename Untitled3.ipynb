```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from google.colab import files
from sklearn.pipeline import make_pipeline

```

```python
data = np.load('ihb.npy')
data = np.nan_to_num(data)
```

```python
print(len(data))
```

```python
def extract_features(data):
    reshaped_data = data.reshape(data.shape[0], -1)  # Преобразуем в двумерный массив
    return reshaped_data
features = extract_features(data)
```

```python
kmeans = KMeans(n_clusters=20, random_state=42)  # 20 кластеров, по числу людей
kmeans_labels = kmeans.fit_predict(features)

# Используем kmeans_labels как целевые метки для обучения
y = kmeans_labels
```

```python
common_brain_region_data = data[:, :, 0]
y = np.array([i // 16 for i in range(320)])
```

```python
# Стандартизация и кластеризация с использованием KMeans
scaler = StandardScaler()
kmeans = KMeans(n_clusters=10, random_state=42)

# Создание пайплайна
model = make_pipeline(scaler, kmeans)
model.fit(common_brain_region_data)
```

```python
# Получение расстояний до центров кластеров
cluster_distances = model.named_steps['kmeans'].transform(common_brain_region_data)

# Инициализация маркировки и индексов
labeling = np.zeros(len(data), dtype=int)
leftover_indexes = np.arange(len(data))

# Определите количество кластеров
num_clusters = model.named_steps['kmeans'].n_clusters

# Жадный алгоритм для выбора объектов
for i in range(num_clusters):  # Используйте num_clusters вместо 20
    distances_from_current_cluster_center = cluster_distances[:, i]
    if len(distances_from_current_cluster_center) > 16:
        top16 = np.argpartition(distances_from_current_cluster_center, 16)[:16]
        labeling[leftover_indexes[top16]] = i
        cluster_distances = np.delete(cluster_distances, top16, axis=0)
        leftover_indexes = np.delete(leftover_indexes, top16)
    else:
        labeling[leftover_indexes] = i
        break
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, y, test_size=0.2, random_state=42)
```

```python
model = Sequential()
model.add(LSTM(128, input_shape=(10, 246), return_sequences=True))
model.add(Dropout(0.3))
model.add(LSTM(64, return_sequences=False))
model.add(Dropout(0.3))
model.add(Dense(32, activation='relu'))
model.add(Dense(20, activation='softmax'))  # 20 классов, по числу людей

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Обучение модели
model.fit(X_train, y_train, epochs=20, batch_size=16, validation_data=(X_test, y_test))

# Предсказания на всех данных (а не только на тестовых)
predictions = model.predict(data)
predicted_classes = np.argmax(predictions, axis=1)
ari_score = adjusted_rand_score(y, predicted_classes)
print(f'Adjusted Rand Score: {ari_score}')
predictions = model.predict(X_test)
predicted_classes = np.argmax(predictions, axis=1)

# Вычисление ARI для оценки качества предсказаний
ari_score = adjusted_rand_score(y_test, predicted_classes)
print(f'Adjusted Rand  for prediction: {ari_score}')
```

```python
submission = pd.DataFrame({
    'prediction': predicted_classes.flatten()  # Преобразуем в одномерный массив
})

# Сохраняем предсказания в файл
submission.to_csv('submission.csv', index=False)

# Скачивание файла на локальный компьютер через Google Colab
from google.colab import files
files.download('submission.csv')
```

```python
import matplotlib.pyplot as plt
def plot_clusters(features, labels):
    pca = PCA(n_components=2)  # Уменьшаем размерность до 2
    reduced_features = pca.fit_transform(features)

    plt.figure(figsize=(10, 8))
    plt.scatter(reduced_features[:, 0], reduced_features[:, 1], c=labels, cmap='viridis', marker='o', alpha=0.6)
    plt.title('KMeans Clustering of Brain Data')
    plt.xlabel('PCA Component 1')
    plt.ylabel('PCA Component 2')
    plt.colorbar(label='Cluster Label')
    plt.show()

# Построим график кластеров
plot_clusters(features, kmeans_labels)
```
